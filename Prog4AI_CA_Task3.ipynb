#### Part 1

```python
#Find average temperature for each location:
avg_temp_location = data.groupby('location_name')['temperature_celsius'].mean().reset_index()
```

Finding the 5 Hottest locations by average temperature in degrees celsius

```python
HottestV = avg_temp_location.sort_values(by='temperature_celsius', ascending=False).head(5)
HottestV.head()
```

Finding the 5 coldest locations by average temperture in degrees celsius

```python
ColdestV = avg_temp_location.sort_values(by='temperature_celsius', ascending=True).head(5)
ColdestV.head()
```

```python
#Find average rainfall of each location in milimetres
avg_precip_location = data.groupby('location_name')['precip_mm'].mean().reset_index()
WettestV = avg_precip_location.sort_values(by='precip_mm', ascending=False).head(5)
```

The top 5 locations with the highest precipitation in mm:

```python
WettestV.head()
```

#### Part 2 - Average, Max and Min temperatures grouped by country

```python
region_stats = data.groupby('country').agg(average_temp=('temperature_celsius', 'mean'),
                                           max_temp=('temperature_celsius', 'max'),
                                           min_temp=('temperature_celsius', 'min')).reset_index()
region_stats.head(10)
```

#### Part 3 - Graphs

```python
data_ireland = data.loc[data['country'] == 'Ireland'].copy()
data_ireland['last_updated'] = pd.to_datetime(data_ireland['last_updated'])

data_italy = data.loc[data['country'] == 'Italy'].copy()
data_italy['last_updated'] = pd.to_datetime(data_italy['last_updated'])

mean_temp = data['temperature_celsius'].mean() #Global mean temp

plt.figure(figsize=(10,6))
sns.lineplot(data=data_ireland, x='last_updated', y='temperature_celsius',label='Ireland')
sns.lineplot(data=data_italy, x='last_updated', y='temperature_celsius', label='Italy')
plt.axhline(y=mean_temp, color='black', linestyle='--', label=f'Global Mean Temp = {mean_temp:.2f}°C')

plt.title("Temperatures in Ireland and Italy during 2024")
plt.ylabel("Temperature (°C)")
plt.ylim(0,40)
plt.xlabel("Year - Month")
plt.legend()
plt.show()
```

```python
mean_humidity = data['humidity'].mean()
data['humidity_level'] = np.where(data['humidity'] > mean_humidity, 'Above Average', 'Below Average')

plt.figure(figsize=(10,6))
sns.histplot(data=data, x='temperature_celsius', hue='humidity_level', bins=20,multiple="dodge")
plt.title("Global Temperature Distribution by Humidity Level (Above and Below Average)")
plt.xlabel("Temperature (°C)")
plt.show()
```

```python
country_data = data.groupby('country').agg(
    max_temp=('temperature_celsius', 'max'),
    latitude=('latitude', 'mean')).reset_index()
topXV = country_data.nlargest(15, 'max_temp')

plt.figure(figsize=(10,6))
sns.scatterplot(topXV,x="max_temp",y="latitude",hue="country", palette='tab10')
plt.title("Higest recorded temperature (2024) vs Latitude for Top 15 Hottest Countries")
plt.xlabel("Temperature (°C)")
plt.ylabel("Latitude")
plt.xlim(40,55)
plt.ylim(0,45)
plt.show()
```
